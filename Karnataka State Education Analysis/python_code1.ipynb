```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('town.csv.csv')
x=[]
x1=[]
x2=[]
y=[]
temp=0
for area,number in zip(df.Area_Name.iteritems(),df.Literate_Persons.iteritems()):
	if area[1] not in x:
		x.append(area[1])
		y.append(number[1])
		
# p,q in df["Area_Name:],df["Total - Persons"] does not work, zip it
for p,q in zip(df["Area_Name"],df["Total_Persons"]):
	if p not in x2:
		x2.append(p)
		x1.append(q)	
print(len(x1))
for i in range(len(y)):
	y[i]=y[i]*100/x1[i]
x2=np.arange(len(x))
plt.bar(x2,y)
plt.xticks(x2,x,rotation=90)
plt.title('percentage of literate people')
plt.xlabel('District name')
plt.show()
```
