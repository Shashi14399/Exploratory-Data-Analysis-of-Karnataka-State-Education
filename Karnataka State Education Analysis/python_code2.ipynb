```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df =  pd.read_csv('Town-wise-education - Karnataka.csv')
```

```python
litf = df["Literate - Females"]
litf1 = []
for i in range(len(litf)):
    litf1.append(litf[i])
#print(litf1)
```

```python
litf2=[]
for i in range(0,811,29):
    litf2.append(litf1[i])
print(litf2)
```

```python
areaname = df["Area Name"]
areaname1=[]
for i in range(len(areaname)):
    areaname1.append(areaname[i])
#print(areaname1)
```

```python
areaname2 = []
for i in range(0,811,29):
    areaname2.append(areaname1[i])
print(areaname2)
```

```python
tf=df["Total - Females"]
tf1=[]
for i in range(len(tf)):
    tf1.append(tf[i])
#print(tf1)
```

```python
tf2=[]
for i in range(0,811,29):
    tf2.append(tf1[i])
print(tf2)
```

```python
lfp=[]
for i in range(len(litf2)):
    lfp.append((litf2[i]/tf2[i])*100)
print(lfp)
```

```python
lfp1=[]
for i in lfp:
    x = int(i)
    lfp1.append(x)
print(lfp1)
    
```

```python
plt.bar(areaname2,lfp1)
plt.title('Female literacy')
plt.xticks(rotation=90)
plt.xlabel('District Name')
plt.ylabel('Percent of literate females')
plt.show()
```

```python
litm = df["Literate - Males"]
litm1 = []
for i in range(len(litm)):
    litm1.append(litm[i])
#print(litm1)
```

```python
litm2=[]
for i in range(0,811,29):
    litm2.append(litm1[i])
print(litm2)
```

```python
tm=df["Total - Males"]
tm1=[]
for i in range(len(tm)):
    tm1.append(tm[i])
#print(tm1)
```

```python
tm2=[]
for i in range(0,811,29):
    tm2.append(tm1[i])
print(tm2)
```

```python
lmp=[]
for i in range(len(litm2)):
    lmp.append((litm2[i]/tm2[i])*100)
#print(lmp)
```

```python
lfp1=[]
for i in lfp:
    x = int(i)
    lfp1.append(x)
print(lfp1)

lmp1=[]
for i in lmp:
    x = int(i)
    lmp1.append(x)
print(lmp1)
    
```

```python
plt.bar(areaname2,lmp1)
plt.title('Male literacy')
plt.xticks(rotation=90)
plt.xlabel('District Name')
plt.ylabel('Percent of literate males')
plt.show()

```

```python
diff = []
for i in range(len(lmp)):
    diff.append(lmp1[i]-lfp1[i])
print(diff)
```

```python
plt.bar(areaname2,diff)
plt.xticks(rotation=90)

plt.xlabel('District Name')
plt.ylabel('Percent Difference in literacy of males and females ')
plt.show()
```

```python
lp = df["Literate - Persons"]
lp1 =[]
for i in range(len(lp)):
    lp1.append(lp[i])
lp2=[]
for i in range(27,811,29):
    lp2.append(lp1[i])
print(lp2)
```

```python
tp=df["Total - Persons"]
tp1 = list(tp)
tp2=[]
for i in range(27,811,29):
    tp2.append(tp1[i])
print(tp2)
```

```python
perc=[]
for i in range(len(tp2)):
    perc.append(int((lp2[i]/tp2[i])*100))
print(perc)
```

```python
plt.bar(areaname2,perc)
plt.xticks(rotation=90)
plt.title('Litercy of senior citizns')
plt.xlabel('District name')
plt.ylabel('percent literacy of senior citizens')
plt.show()
```
